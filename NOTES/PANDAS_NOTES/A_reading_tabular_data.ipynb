# READING TABULAR DATA

```python
import pandas as pd
```

```python
pd.read_table("pandas-videos-master/data/chipotle.tsv")
```

```python
movies=pd.read_table("pandas-videos-master/data/u.user")
movies.head()
```

## Above df uses | as  seperator.To solve it we will use the sep parameter of read_table

```python
pd.read_table("pandas-videos-master/data/u.user",sep="|")
```

## Now the first row as can be seen is not the header row.  To correct it we will use the header parameter

```python
df=pd.read_table("pandas-videos-master/data/u.user",sep="|",header=None)
df.head()
```

## Assigning appropriate col names

```python
col_names=['user_id','age','gender','occupation','zip_code']
user=pd.read_table("pandas-videos-master/data/u.user",sep="|",header=None,names=col_names)
user.head()
```

## or

```python
col_names=['user_id','age','gender','occupation','zip_code']
user=pd.read_table("pandas-videos-master/data/u.user",sep="|",header=None)
user.columns=col_names
user.head()
```
