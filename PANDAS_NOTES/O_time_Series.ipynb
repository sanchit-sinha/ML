# TIME_SERIES

```python
import pandas as pd
```

```python
ufo=pd.read_csv("pandas-videos-master/data/ufo.csv")
ufo.head()
```

```python
ufo.dtypes
```

## CONERTING TYPE TO DATETIME

```python
ufo["Time"] = pd.to_datetime(ufo.Time)
ufo.head()
```

```python
ufo.dtypes
```

## ATTRIBUTES OF DATE-TIME

```python
ufo["days"]=ufo.Time.dt.day_name()
ufo.head()
```

## FILTERING ROWS WITH TIME

```python
date=pd.to_datetime("1/2/1999")
```

```python
ufo.loc[ufo.Time >= date,:]
```

## TIMESTAMP

```python
dif=(ufo.Time.max()-ufo.Time.min())
dif
```

```python
dif.days
```

## PLOTTING YEAR VS COUNTS

```python
%matplotlib inline
```

```python
ufo["Year"]=ufo.Time.dt.year
```

```python
ufo.head()
```

```python
ufo.Year.value_counts()
```

```python
ufo.Year.value_counts().sort_index()
```

```python
ufo.Year.value_counts().sort_index().plot(kind="line")
```
