# USING MAP,APPLY-MAP,APPLY

```python
import pandas as pd
```

```python
train=pd.read_csv("pandas-videos-master/data/titanic_train.csv")
train.head()
```

## USING MAP(NOT GENERALLY USED)

```python
train["Sex_Male"]=train.Sex.map({"female":0,"male":1})
train.head()
```

## APPLY IS USED FOR SERIES AS WELL AS DATAFRAME

#### IT IS USED TO APPLY A FUNCTION TO ALL-----------

## USING IT WITH SERIES

```python
train["Name_len"]=train.Name.apply(len)
```

```python
train.head()
```

### APPLYING CEILING FUNCTION FROM NUMPY

```python
import numpy as np
```

```python
train["Fare_Ceil"]=train.Fare.apply(np.ceil)
```

```python
train.head()
```

### GETTING FIRST NAMES OF Name COLUMN

```python
train["first_name"] =train.Name.str.split(",")
train.first_name.head()
```

#### WRITING AOUR OWN FUNCTION TO GET THE FIRST PART FROM THE LIST OF , SEPERATED NAMES

```python
def get_first_name(list,position):
    return list[position]
```

```python
train["first_name"] =train.Name.str.split(",").apply(get_first_name,position=0)
train.first_name.head()
```

## USING APPLY FOR DATAFRAME

```python
drinks=pd.read_csv("pandas-videos-master/data/drinks.csv")
drinks.head()
```

```python
drinks.loc[0:4,"beer_servings":"wine_servings"].apply(max,axis=0)
```

```python
drinks.loc[0:4,"beer_servings":"wine_servings"].apply(max,axis=1)
```

```python
import numpy as np
```

```python
drinks.loc[:,"beer_servings":"wine_servings"].apply(np.argmax,axis=1)
```
